```python
import pandas as pd
```

```python
import json
```

```python
with open('/content/t20_wc_match_results.json') as f:
  data=json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
```

```python
df_match.head(2)
```

```python
match_ids_dict={}
for index,row in df_match.iterrows():
  key1=row['team1'] + ' Vs ' + row['team2']
  key2=row['team2'] + ' Vs ' + row['team1']
  match_ids_dict[key1]=row['match_id']
  match_ids_dict[key2]=row['match_id']



```

```python
match_ids_dict
```

```python
df_match.to_csv('/content/sample_data/dim_match_summary.csv', index = False)
```

Batting summary

```python
with open('/content/t20_wc_batting_summary.json') as f:
  data=json.load(f)
data
all_records=[]
for rec in data:
  all_records.extend(rec['battingSummary'])
df_batting=pd.DataFrame(all_records)
df_batting.head(10)
```

```python
df_batting['out/notout']=df_batting.dismissal.apply(lambda x:'out' if len(x)>0 else 'notout')
```

```python
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
```

```python
df_batting.head(11)
```

```python
df_batting['batsmanName']=df_batting.batsmanName.apply(lambda x: x.replace('†',''))
df_batting['batsmanName']=df_batting.batsmanName.apply(lambda x: x.replace('\xa0',''))
```

```python
df_batting.head(150)
```

```python
match_ids_dict['Sri Lanka Vs Netherlands']
```

```python
df_batting['match_id']=df_batting['match'].map(match_ids_dict)
```

```python
df_batting
```

```python
df_batting.shape
```

```python
df_batting.to_csv('/content/sample_data/fact_bating_summary.csv', index = False)
```
